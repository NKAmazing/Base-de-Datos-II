```python
import pandas as pd
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
df_full = pd.read_csv('hotel_bookings.csv').dropna()
df_full = pd.DataFrame(df_full)
df = df_full[['hotel', 'arrival_date_month', 'arrival_date_year']]
df = df.rename(columns={"hotel": "Hotel", "arrival_date_month": "Month", "arrival_date_year": "Year"})
data = df.groupby(["Hotel", "Year"]).count().plot(kind="bar", figsize=(10,8))
plot.ylabel("Number of Bookings")
plot.title("Number of Bookings per Hotel and per Year")
plot.xticks(rotation=45)
plot.xlabel("Hotel and Year")
#Este anÃ¡lisis me trae la cantidad de reservas por aÃ±o, para cada hotel.
df_full
df
```
