## Ejercicio 1

- Descomprimir el archivo clima.tgz de la carpeta data
- Convertir cada archivo a Dataframe y mostrarlo en pantalla

```python
import tarfile as t
import pandas as pd

path = r"C:\Users\Nicolas\Proyectos Programacion\Base-de-Datos-II\base-de-datos-ii\data\clima.tgz"
new_path = r"C:\Users\Nicolas\Proyectos Programacion\Base-de-Datos-II\base-de-datos-ii\data\TP1/"

item = t.open(path, "r:*")

item.extractall(new_path)

name_file = item.getnames()

item.close()

for i in name_file:
    file = new_path + i
    df = pd.read_csv(file, delimiter= "\t", encoding='latin-1', header=[0,1])
    print(i, "/n", df)
```

## Ejercicio 2

- Obtener de la API de <a href="https://wikimedia.org/api/rest_v1/#/Pageviews%20data">Wikimedia </a> la cantidad de visitas en este año a la página Albert Einstein por día.
- Mostrar el Dataframe obtenido por pantalla

```python
import requests
import pandas as pd

url = "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia.org/all-access/all-agents/Albert_Einstein/daily/20220101/20221229"
header = {'user-agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36"}


r = requests.get(url, headers=header)

# Convertir JSON a diccionario

data = r.json()

df = pd.DataFrame(data["items"])

df
```

## Ejercicio 4

- Obtener de la API <a href="https://date.nager.at/swagger/index.html">Nager.Date</a> las fiestas nacionales argentinas para el 2021
- Mostrar el dataframe resultante por pantalla

```python
import requests
import pandas as pd

countryCode = "AR"
year = 2022

url = f"https://date.nager.at/api/v3/PublicHolidays/{year}/{countryCode}"

r = requests.get(url)
data = r.json()

df = pd.DataFrame(data)

df
```
