Es una mala representacion porque el grafico no tiene proporcion de los tamaños de cada uno.

```python
import matplotlib.pyplot as plt
%matplotlib inline

labels = ["Babar", "Rohit", "Shakib", "Warner", "Williamson", "Rahmat Shah", "Root", "Kusal Perera", "Pooran", "Du Plessis"]
sizes = [24.51, 29.05, 28.25, 25.02, 30.23, 14.8, 19.07, 18.16, 20.01, 21.06]
explode = [0, 0.2, 0, 0, 0.2, 0, 0, 0, 0, 0]
fig1, a = plt.subplots()
a.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',)
a.axis('equal')
plt.show()
```

```python
print("TRENT BOULT")
labels = '% OF WKTS', '% OF WKTS IN WINS', '% OF WKTS AT HOME', '% OF WKTS IN HOME WINS'
sizes = [23, 24, 26, 27]
explode = (0, 0., 0, 0)
fig1, a = plt.subplots()
a.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',)
a.axis('equal')
plt.show()
```
